# Marine Heatwaves in the Gulf of Naples (2014–2024)

This notebook analyses marine heatwaves in the Gulf of Naples using
a pre-computed event catalogue and yearly summary generated from
NOAA OISST daily sea surface temperature (SST) data.

The detection follows the Hobday et al. (2016) definition:

- A marine heatwave (MHW) occurs when SST exceeds the **90th percentile**
  of the local, calendar-day climatology
- The exceedance must last for **at least 5 consecutive days**

The underlying Python script:
- loads multi-year SST from netCDF files,
- computes a baseline climatology (2014–2019),
- detects MHW events,
- exports:
  - a detailed event table (`marine_heatwaves_events.csv`)
  - a yearly summary (`marine_heatwaves_yearly_summary.csv`)
  - a time-series plot with shaded MHWs.


## Study Area

The analysis focuses on the Gulf of Naples (Tyrrhenian Sea, Central Mediterranean).

Bounding box:
- Latitude: 40.3°N → 41.1°N  
- Longitude: 13.8°E → 14.8°E  

NOAA OISST provides NaN over land pixels, so only ocean data is used.


## Methodology

Marine heatwaves (MHWs) follow Hobday et al. (2016):

1. SST > 90th percentile of the daily climatology  
2. Event lasts ≥ 5 consecutive days  
3. Baseline climatology period used: **2014–2019**  
   (earliest available years in the dataset)

The main script:
- loads multiple OISST netCDF files
- extracts the Gulf of Naples region
- computes daily climatology and threshold
- detects marine heatwave events
- saves the event table and yearly summary


## Data Used in This Notebook

This notebook does **not** load raw netCDF files.  
It uses the CSV outputs produced by the detector script:

- `tables/marine_heatwaves_events.csv`  
- `tables/marine_heatwaves_yearly_summary.csv`  
- `plots/sst_mhw_timeseries.png`


```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure plots show inside the notebook
%matplotlib inline

# Paths to the tables generated by your main.py script
events_path = "tables/marine_heatwaves_events.csv"
summary_path = "tables/marine_heatwaves_yearly_summary.csv"

# Load CSVs
events = pd.read_csv(events_path, parse_dates=["start", "end"])
summary = pd.read_csv(summary_path)

events.head(), summary.head()
```

```python
print("Number of events:", len(events))
print("Years covered in summary:", summary["year"].min(), "→", summary["year"].max())

summary
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Top: number of events per year
axes[0].bar(summary["year"], summary["n_events"])
axes[0].set_ylabel("Number of MHW events")
axes[0].set_title("Marine heatwave frequency per year")

# Bottom: total MHW days per year
axes[1].bar(summary["year"], summary["total_mhw_days"])
axes[1].set_ylabel("Total MHW days")
axes[1].set_xlabel("Year")
axes[1].set_title("Total days spent in marine heatwave per year")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4))

ax.plot(summary["year"], summary["max_intensity_degC"], marker="o", label="Max intensity")
ax.plot(summary["year"], summary["mean_intensity_degC"], marker="s", label="Mean intensity")

ax.set_xlabel("Year")
ax.set_ylabel("Intensity (°C above threshold)")
ax.set_title("Marine heatwave intensity in the Gulf of Naples")
ax.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Interpretation of marine heatwave trends (2014–2024)

From the yearly summary, several clear patterns emerge:

1. **Increasing heatwave persistence**  
   Total days in marine heatwave per year increase dramatically over time.
   Early years (2014–2016) show fewer than ~100 MHW days, while recent years
   such as **2022** and especially **2024** exceed 130 and 200 MHW days,
   respectively. In 2024, the Gulf of Naples spent **more than half the year**
   in a heatwave state.

2. **Growing intensity of extremes**  
   Maximum anomalies rise from < 1°C in the mid-2010s to around **2–2.3°C**
   in 2022–2023, indicating that peak events are becoming substantially warmer
   relative to the historical threshold.

3. **Longer events and seasonal shifts**  
   The longest events extend from ~10–20 days in the early years to
   **45–63 days** after 2022. Marine heatwaves increasingly span multiple
   months and are no longer confined to peak summer: winter and shoulder-season
   events appear more frequently, consistent with a warming baseline climate.

4. **Ecological implications**  
   Persistent marine heatwaves of this magnitude are known to affect
   plankton phenology, seagrass meadows, benthic communities and fish
   distribution in the Mediterranean. The results suggest that the Gulf of
   Naples has transitioned from occasional extremes to a regime where long,
   recurrent marine heatwaves are a dominant feature of the thermal environment.


```python
import numpy as np

years = summary["year"].values
days = summary["total_mhw_days"].values

coef = np.polyfit(years, days, 1)
trend = np.polyval(coef, years)

print(f"Linear trend: {coef[0]:.2f} MHW days per year")

plt.figure(figsize=(8,4))
plt.scatter(years, days, label="Total MHW days")
plt.plot(years, trend, color="red", label="Linear trend")
plt.xlabel("Year")
plt.ylabel("Total MHW days")
plt.title("Trend in marine heatwave days (Gulf of Naples)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
